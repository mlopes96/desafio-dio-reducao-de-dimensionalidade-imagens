# Redução de Dimensionalidade em Imagens
## Desafio Formação Machine Learning DIO
Seguindo o exemplo do algoritmo apresentado em aula, realize a implementação em Python para transformar uma imagem colorida para níveis de cinza (0 a 255) e para binarizada (0 e 255), preto e branco. Observação: O desafio deve ser realizado sem a utilização de bibliotecas prontas para conversão de imagens.

```python
# Inicia importando as bibliotecas
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive

# Faz a conexão com o Google Drive.
drive.mount('/content/drive')

# Caminho para a pasta onde estão as imagens
path = f'/content/drive/MyDrive/Colab Notebooks/ ReduçãoDimensionalidadeImagens'

# Caminho para a imagem
image_path = f'{path}/Lenna.png'

# Carrega a imagem como uma matriz NumPy
imagem = plt.imread(image_path)

# Exibe a imagem carregada
plt.imshow(imagem)
plt.axis('off')
plt.show()
```

```python
# Confere se a imagem tem 3 canais de cores (RGB)
if imagem.shape[2] == 3:
    # Converte a imagem para escala de cinza
    imagem_cinza = np.dot(imagem[...,:3], [0.2989, 0.5870, 0.1140])
else:
    print('A imagem não tem 3 canais de cores. A conversão não pôde ser feita.')
    # Armazena a imagem em imagem_cinza
    imagem_cinza = imagem

# Normaliza os valores para o intervalo [0, 255]
imagem_cinza = (imagem_cinza * 255).astype(np.uint8)

# Salva a imagem em escala de cinza
plt.imsave(f'{path}/imagem-em-escala-de-cinza.png', imagem_cinza, cmap='gray')

# Exibe a imagem em escala de cinza
plt.imshow(imagem_cinza, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Define um limite (threshold) para binarização
limite = 128

# Aplica a binarização
imagem_binarizada = np.where(imagem_cinza > limite, 255, 0)

# Salva a imagem binarizada
plt.imsave(f'{path}/imagem-binarizada.png', imagem_binarizada, cmap='gray')

# Exibe a imagem binarizada
plt.imshow(imagem_binarizada, cmap='gray')
plt.axis('off')
plt.show()
```
